```python
da = da.astype('category')
```

```python
da.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in da.columns:
  da[column]=le.fit_transform(da[column])
```

```python
da.head()
```

# Storing feature matrix in X(independent variable) and Responses(Target) in Vector Y(dependent variable)

```python
Y = da.drop('class',axis=3)
X = da['class']
```

```python
X
```

```python
Y
```

# Applying PCA

```python
from sklearn.decomposition import PCA
pca1 = PCA(n_components=5)
pca_fit = pca1.fit_transform(Y)
```

```python
pca1.explained_variance_ratio_
```

```python
from sklearn.model_selection import train_test_split
Y_train,Y_test,X_train,X_test = train_test_split(pca_fit,X,test_size =0.20,random_state=42)
```

```python
sum(pca1.explained_variance_ratio_)
```

# Splitting the dataset into training set and test set

```python
X_train
```

```python
Y_train
```

# Importing the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Training the model

```python
lr= LogisticRegression ()
lr.fit(Y_train,X_train)
rf=RandomForestClassifier()
rf.fit(Y_train,X_train)
knn =KNeighborsClassifier()
knn.fit(Y_train,X_train)
```

# Prediction on test data

```python
X_pred1 = lr.predict(Y_test)
X_pred2 = knn.predict(Y_test)


X_pred5 = rf.predict(Y_test)
```

# Evaluating the Algorithm

```python
from sklearn.metrics import accuracy_score
```

```python
print("ACC LR",accuracy_score(X_test,X_pred1))
print("ACC KNN",accuracy_score(X_test,X_pred2))
print("ACC RF",accuracy_score(X_test,X_pred5))
```

```python
final_da = pd.DataFrame({'Models':['LR','KNN','RN',],
              'ACC':[accuracy_score(X_test,X_pred1)*100,
                      accuracy_score(X_test,X_pred2)*100,

                      accuracy_score(X_test,X_pred5)*100]})
```
